## 利用爬蟲技術，取得台灣彩券，最近一期，各獎開獎結果
### https://www.taiwanlottery.com.tw/




```python
import requests
from bs4 import BeautifulSoup

# 目標網站
url = 'https://www.taiwanlottery.com.tw/'
r = requests.get(url) 
sp = BeautifulSoup(r.content, features='html')
```

```python
def Top_Dollar():
    top_dollar_n = sp.find_all('div', class_="top_dollar") #預測頭獎金額的區塊
#     for i in range(0,2):
#         print(top_dollar_n[i].text,end=' ')
    print("最新頭彩預估金額:")
    print("-"*25)
    print("威力彩$ :",top_dollar_n[0].text,end=' 台幣'+"\n")
    print("大樂透$ :",top_dollar_n[1].text,end=' 台幣'+"\n"+"-"*25+"\n")
    
def BINGO_BINGO():
    # BINGO BINGO
    A="-"*95
    datas_bingo = sp.find('div', class_="contents_box01")
    # 開獎期數
    title = datas_bingo.find('span',class_="font_black15").text
    print("BINGO BINGO:"+"\n"+title+":"+"\n"+A,)
    # 開獎號碼
    nums = datas_bingo.find_all('div',class_="ball_tx ball_yellow")
    print("開出黃色球號:",end=' ')
    for i in range(0,20):
        print(nums[i].text,end=' ')
    print()
    # 超級獎號
    nums = datas_bingo.find_all('div',class_="ball_red")
    print("開出超級獎號-紅色球號:",end=' ')
    for i in range(0,1):
        print(nums[i].text,end=' '+"號"+"\n")
    # 猜大小 & 猜單雙
    nums = datas_bingo.find_all('div',class_="ball_blue_BB1")
    print("猜大小:",end=' ')
    for i in range(0,1):
        print(nums[i].text,end=' ')
    nums = datas_bingo.find_all('div',class_="ball_blue_BB2")
    print("\n"+"猜單雙:",end=' ')
    for i in range(0,1):
        print(nums[i].text,end=' '+"\n")
    
    
def Lotto_22():   
    A="-"*60
    print(A)
    datas_22 = sp.find('div', class_="contents_box06")
    # 開獎期數
    title = datas_22.find('span',class_="font_black15").text
    print("雙贏彩:"+"\n"+title+"\n"+A)
    # 開獎號碼
    nums = datas_22.find_all('div',class_="ball_tx ball_blue")
    print("開出順序:",end=' ')
    for i in range(0,12):
        print(nums[i].text,end=' ')
    print()
    # 開獎大小順序
    print("大小順序:",end=' ')
    for i in range(12,24):
        print(nums[i].text,end=' ')
    print("\n"+A)
    
def Superlotto():
    A="-"*50
    B="\n"
    datas_S = sp.find_all('div', class_="contents_box02") #contents_box02 類別有四塊
    datas_SL = datas_S[0]                                      #威力彩區塊
    title_SL = datas_SL.find('span', class_="font_black15") 
    print("威力彩:"+B+title_SL.text+"\n"+A)
    num_SL = datas_SL.find_all('div', class_="ball_tx ball_green")  
    print("開出順序：",end=" ")
    for i in range(0,6):
        if i == 5: 
            print(num_SL[i].text)
            break
        print(num_SL[i].text,end=" ")
    print("大小順序：",end=" ")
    for i in range(6,12):
        if i == 11: 
            print(num_SL[i].text)
            break
        print(num_SL[i].text,end=" ")
    second_SL = datas_SL.find('div', class_="ball_red") 
    print("第二區：", second_SL.text,end=" "+"\n"+A+B)
    
def Superlotto_38():
    A="-"*50
    B="\n"
    datas_S = sp.find_all('div', class_="contents_box02")
    datas_SL38 = datas_S[1]
    title_SL38 = datas_SL38.find('span', class_="font_black15")  
    print("38樂合彩:"+B+title_SL38.text,end="　"+B+A)
    print()
    num_SL38 = datas_SL38.find_all('div', class_="ball_tx ball_green") 
    print("開出順序：",end=" ")
    for i in range(0,6):
        if i == 5: 
            print(num_SL38[i].text)
            break
        print(num_SL38[i].text,end=" ")
    print("大小順序：",end=" ")
    for i in range(6,12):
        if i == 11: 
            print(num_SL38[i].text)
            break
        print(num_SL38[i].text,end=" ")
    print(A)

def Lotto_649():
    A="-"*50
    B="\n"
    datas_S = sp.find_all('div', class_="contents_box02")
    datas_LT649 = datas_S[2]
    #print(datas_LT649)
    title_LT649 = datas_LT649.find('span', class_="font_black15")  
    print("大樂透:"+B+title_LT649.text+B+A,end="　"+B)
    num_LT649 = datas_LT649.find_all('div', class_="ball_tx ball_yellow")  
    print("開出順序：",end=" ")
    for i in range(0,6):
        if i == 5: 
            print(num_LT649[i].text)
            break
        print(num_LT649[i].text,end=", ")
    print("大小順序：",end=" ")
    for i in range(6,12):
        if i == 11: 
            print(num_LT649[i].text)
            break
        print(num_LT649[i].text,end=", ")
    second_LT649 = datas_LT649.find('div', class_="ball_red")
    print("第二區：", second_LT649.text,end=" "+B+A+B)
    
def Lotto_49():
    A="-"*50
    B="\n"
    datas_S = sp.find_all('div', class_="contents_box02")
    datas_LT49 = datas_S[3]
    title_LT49 = datas_LT49.find('span', class_="font_black15") 
    print("49樂合彩:"+B+title_LT49.text,end="　"+B+A+B)
    num_LT49 = datas_LT49.find_all('div', class_="ball_tx ball_yellow")  
    print("開出順序：",end=" ")
    for i in range(0,6):
        if i == 5: 
            print(num_LT49[i].text)
            break
        print(num_LT49[i].text,end=" ")
    print("大小順序：",end=" ")
    for i in range(6,12):
        if i == 11: 
            print(num_LT49[i].text)
            break
        print(num_LT49[i].text,end=" ")
    print(A)
```

```python
# 主程式
print("台灣樂透彩開獎結果:"+"\n"+"https://www.taiwanlottery.com.tw/"+"\n"+"↓"*50)
print()
Top_Dollar()
BINGO_BINGO()
Lotto_22()
Superlotto()
Superlotto_38()
Lotto_649()
Lotto_49()
```
